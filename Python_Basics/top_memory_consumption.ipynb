# Top Memory Consuming process 

```python
# Importing subprocess library to perform process related operation
import subprocess

print("import successful")
```

```python
# Getting the list of process currently running on the system
output = subprocess.check_output("tasklist", shell = True).decode()

# Printing the list
output
```

```python
# Printing the type of the output
print(type(output))
```

```python
# Excluding the  starting lines since of not much use
output_lines = output.split("\r\n")[3:]

# Print 
output_lines
```

```python
# Split the output line
output_split = [line.split() for line in output_lines if line.strip()]

# Print
output_split
```

```python
# Function to get the  highest memory consuming process
def get_mem(l):
    mem_str = "".join(l[-2].split(","))
    return int(mem_str)
```

```python
# Just a random pass to the function
get_mem(['System', '4', 'Services', '0', '2,176', 'K'])
```

```python
# From the reversed  list of the process iterate to find the top 5 highest memory consuming process
for line in sorted(output_split, key = get_mem, reverse = True)[:5]:
    print(line[0], "\t\t", line[-2], "K")
```
